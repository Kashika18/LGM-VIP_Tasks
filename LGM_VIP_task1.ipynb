#Image to Pencil Sketch with Python:



image.png

```python
import cv2
```

#We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. 

```python
#reading the image
image = cv2.imread("sketch.jpg")

from google.colab.patches import cv2_imshow

cv2_imshow(image)
```

```python
#converting the original image to greyscale

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

# Then the next thing to do is invert the grayscale image
 also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details.

```python
#inverting the new grayscale image
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
```

```python
#blurring the image by using the Gaussian Function in OpenCV

blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
cv2_imshow(blurred)
```

```python
#inverting the blurred image
inverted_blurred = 225- blurred
cv2_imshow(inverted_blurred)
```

# Then we can finally create the pencil sketch 
by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

```python
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=200.0)
cv2_imshow(pencil_sketch)
```
